```python
import filecmp
  
f1 = "/Users/harman/Downloads/2_g.txt"
f2 = "/Users/harman/Downloads/3_g.txt"
  
# shallow comparison
result = filecmp.cmp(f1, f2)
print(result)
# deep comparison
result = filecmp.cmp(f1, f2, shallow=False)
print(result)
```

```python

# reading files
f1 = open("/Users/harman/Downloads/2_g.txt", "r")  
f2 = open("/Users/harman/Downloads/3_g.txt", "r")  
  
i = 0
numlines = 0
wrong = 0
for line1 in f1:
    i += 1
    for line2 in f2:
          
        # matching line1 from both files
        if line1 != line2:  
            print("Line ", i, ":")
            # else print that line from both files
            print("\tFile 1:", line1, end='')
            print("\tFile 2:", line2, end='')
            wrong+=1
        numlines+=1
        break
        
  
print("wrong", wrong)
print("total", numlines)
# closing files
f1.close()                                       
f2.close()            
```
